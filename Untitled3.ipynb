```python
%matplotlib inline
import matplotlib.pyplot as plt
from pykrx import stock

class code():
    def __init__(self, start_date='2019-08-30', end_date='2023-08-30', code_lst=['058730', '001470']):
        self.start_date = start_date
        self.end_date = end_date
        self.code_lst = code_lst
        self.ohlcv_dict = {}
        self.cap_shares_dict = {}

        self.get_info()


    def get_info(self):
        for code in self.code_lst:
            ### Get OHLCV
            self.ohlcv_dict[code] = stock.get_market_ohlcv_by_date(self.start_date, self.end_date, code)
            # Colume Handling
            self.ohlcv_dict[code] = self.ohlcv_dict[code].reset_index()
            self.ohlcv_dict[code].columns = ['date', 'open', 'high', 'low', 'close', 'volume', 'rate']
            #self.ohlcv_dict[code] = self.ohlcv_dict[code][['date', 'open', 'high', 'low', 'close', 'volume']]

            ### Get CAP, Shares
            self.cap_shares_dict[code] = stock.get_market_cap_by_date(self.start_date, self.end_date, code)
            # Colume Handling
            self.cap_shares_dict[code] = self.cap_shares_dict[code].reset_index()
            self.cap_shares_dict[code].columns = ['date', 'cap', 'volume', 'volume_price', 'shares']

    def price_plot(self):
        fig, ax = plt.subplots(len(self.code_lst), 1, figsize=(8, 4))
        for idx, code in enumerate(self.code_lst):
            # Close Price
            ax[idx].plot(self.ohlcv_dict[code]['date'].tolist(), self.ohlcv_dict[code]['close'].tolist())    #print(self.ohlcv_dict[code])


    def info_plot(self):
        fig, ax = plt.subplots(len(self.code_lst), 2, figsize=(8, 4))
        for idx, code in enumerate(self.code_lst):
            print(self.cap_shares_dict[code])
            # Cap
            ax[idx][0].plot(self.cap_shares_dict[code]['date'].tolist(), self.cap_shares_dict[code]['cap'].tolist())
            # Shares
            ax[idx][1].plot(self.cap_shares_dict[code]['date'].tolist(), self.cap_shares_dict[code]['shares'].tolist())
```

```python
stock = code()
stock.price_plot()
```
